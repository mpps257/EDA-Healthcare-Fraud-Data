## Solution - 1

```python
in_merge_fraud = pd.merge(inpatient_train, fraud_train, on="Provider", how="left")
out_merge_fraud = pd.merge(outpatient_train, fraud_train, on="Provider", how="left")
```

```python
in_merge_fraud.columns
```

```python
out_merge_fraud.columns
```

```python
# Inpatient and Outpatient fraudulent cases  
in_fraud = in_merge_fraud[in_merge_fraud["PotentialFraud"] == "Yes"][["BeneID", "ClaimID", "Provider","AttendingPhysician","InscClaimAmtReimbursed"]]
out_fraud = out_merge_fraud[out_merge_fraud["PotentialFraud"] == "Yes"][["BeneID", "ClaimID", "Provider","AttendingPhysician","InscClaimAmtReimbursed"]]
```

```python
in_fraud.head()
```

```python
out_fraud.head()
```

```python
#Merge the dataframes along the rows and reset the index
fraud = pd.concat([in_fraud, out_fraud], axis=0,ignore_index=True)
fraud.head()
```

```python

top_by_count = (fraud.groupby('AttendingPhysician')
                .size()
                .reset_index(name='FraudulentClaimCount')
                .sort_values(by='FraudulentClaimCount', ascending=False) # Sort in descending order
                .head(20))
top_by_count
```

```python
top_by_count = fraud.groupby("AttendingPhysician").size().sort_values(ascending=False).head(20)
top_by_count
```

```python
top_by_amount = (fraud.groupby('AttendingPhysician')
                 .agg({'InscClaimAmtReimbursed': 'sum'})
                 .sort_values(by='InscClaimAmtReimbursed', ascending=False)
                 .head(20))
top_by_amount
```

```python
top_by_amount = fraud.groupby("AttendingPhysician")["InscClaimAmtReimbursed"].sum().sort_values(ascending=False).head(20)
top_by_amount
```

## Solution 2

```python
statewise_ben = ben_train[['BeneID', 'State']].groupby('State').size().reset_index(name='BeneficiaryCount').sort_values(by='BeneficiaryCount', ascending=False)
statewise_ben
```

```python
total_beneficiaries = statewise_ben['BeneficiaryCount'].sum()
statewise_ben['Percentage'] = (statewise_ben['BeneficiaryCount'] / total_beneficiaries) * 100
statewise_ben
```
